```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mutual_info_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Ridge
from sklearn.metrics import accuracy_score
```

```python
df_full = pd.read_csv("https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-02-car-price/data.csv")
```

```python
df_full.head()
```

```python
columns_to_keep = ['Make', 'Model', 'Year', 'Engine HP', 'Engine Cylinders', 'Transmission Type', 'Vehicle Style', 'highway MPG', 'city mpg', 'MSRP']

df = df_full[columns_to_keep]
df.head()
```

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
df
```

```python
df.rename(columns = {'msrp':'price'}, inplace = True) 
df
```

```python
# Which columns/features have NAs/Nulls?
df.isnull().sum()
```

```python
# Replace NAs by zero
df.fillna(0, inplace=True)
df.isnull().sum()
```

```python
# Question 2: What are the two features that have the biggest correlation in this dataset?
numerical = ['year', 'engine_hp', 'engine_cylinders', 'highway_mpg', 'city_mpg', 'price']
cormat = df[numerical].corr(method='pearson')
```

```python
# Question 3. Which of these variables has the lowest mutual information score?

# Let's create a variable above_average which is 1 if the price is above its mean value and 0 otherwise.
df['above_average'] = (df.price > df.price.mean()).astype(int)
df['above_average']
```

```python
df_for_training = df.copy()
del df_for_training['price']
categorical = ['make', 'model', 'transmission_type', 'vehicle_style']

df_full_train, df_test = train_test_split(df_for_training, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)
y_train = df_train.above_average.values
y_val = df_val.above_average.values
y_test = df_test.above_average.values

del df_train['above_average']
del df_val['above_average']
del df_test['above_average']
```

```python
for category in categorical:
    print(f'{category} : {round(mutual_info_score(df_train[category], y_train),2)}')
```

```python
# 'transmission_type' has the lowest mutual info with 'above average'
```

```python
df_train
```

```python
# Question 4: What accuracy did you get?
numerical = numerical[:-1] # Drop 'price' from features 

scaler = StandardScaler()
#scaler = MinMaxScaler()

X_train_num = df_train[numerical].values
X_train_num = scaler.fit_transform(X_train_num)

ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')
X_train_cat = ohe.fit_transform(df_train[categorical].values)

X_train = np.column_stack([X_train_num, X_train_cat])

X_val_num = df_val[numerical].values
X_val_num = scaler.transform(X_val_num)

X_val_cat = ohe.transform(df_val[categorical].values)

X_val = np.column_stack([X_val_num, X_val_cat])

model = LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_val)
original_accuracy = round(accuracy_score(y_val, y_pred),2)
print(f"Accuracy on the validation dataset: {original_accuracy}")
```

```python
# We get accuracy = 0.95 (nearest among options provided)
```

```python
# Question 5: Which of following feature has the smallest difference?

# print (f'Accuracy with all features: {round(original_accuracy,3)}')

drops = {}

for category in categorical:
    
    c = categorical.copy()
    c.remove(category)

    scaler = StandardScaler() #scaler = MinMaxScaler()
    X_train_num = df_train[numerical].values
    X_train_num = scaler.fit_transform(X_train_num)
    
    ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')
    X_train_cat = ohe.fit_transform(df_train[c].values)
    X_train = np.column_stack([X_train_num, X_train_cat])
    
    X_val_num = df_val[numerical].values
    X_val_num = scaler.transform(X_val_num)
    
    X_val_cat = ohe.transform(df_val[c].values)
    X_val = np.column_stack([X_val_num, X_val_cat])

    model.fit(X_train, y_train)

    y_pred = model.predict_proba(X_val)[:, 1]
    above_avg = y_pred > 0.5
    
    accuracy_without_category = (y_val == above_avg).mean()
    change_in_accuracy = original_accuracy - accuracy_without_category
    
    drops[category] = change_in_accuracy
    
    #print(f'{category} : Accuracy without feature: {round(accuracy_without_category,6)}\t Drop in accuracy: {round(change_in_accuracy, 7)}')

for num in numerical:
    n = numerical.copy()
    n.remove(num)

    scaler = StandardScaler() #scaler = MinMaxScaler()
    X_train_num = df_train[n].values
    X_train_num = scaler.fit_transform(X_train_num)
    
    ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')
    X_train_cat = ohe.fit_transform(df_train[categorical].values)
    X_train = np.column_stack([X_train_num, X_train_cat])
    
    X_val_num = df_val[n].values
    X_val_num = scaler.transform(X_val_num)
    
    X_val_cat = ohe.transform(df_val[categorical].values)
    X_val = np.column_stack([X_val_num, X_val_cat])

    model.fit(X_train, y_train)

    y_pred = model.predict_proba(X_val)[:, 1]
    above_avg = y_pred > 0.5
    
    accuracy_without_category = (y_val == above_avg).mean()
    change_in_accuracy = original_accuracy - accuracy_without_category
    
    drops[num] = change_in_accuracy
    
    #print(f'{num} : Accuracy without feature: {round(accuracy_without_category,6)}\t Drop in accuracy: {round(change_in_accuracy, 7)}')
    
    drops_sorted = dict(sorted(drops.items(), key=lambda x: x[1]))
    drops_sorted
```

```python
# Question 6: Which of these alphas leads to the best RMSE on the validation set?

def rmse(y, y_pred):
    error = y_pred - y
    mse = (error ** 2).mean()
    return np.sqrt(mse)
```

```python
df_for_linreg = df.copy()
del df_for_linreg['above_average']

df_full_train, df_test = train_test_split(df_for_linreg, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)
y_train = np.log1p(df_train.price.values)
y_val = np.log1p(df_val.price.values)
y_test = np.log1p(df_test.price.values)

del df_train['price']
del df_val['price']
del df_test['price']

train_dict = df_train[categorical + numerical].to_dict(orient='records')
# train_dict = df_train[numerical].to_dict(orient='records')


dv = DictVectorizer(sparse=True)
dv.fit(train_dict)
X_train = dv.transform(train_dict)

val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)

for a in [0, 0.01, 0.1, 1, 10]:

    ridge = Ridge(alpha=a,  solver='sag')
    ridge.fit(X_train, y_train)

    y_pred = ridge.predict(X_val)
    train_score = rmse(y_val, y_pred)
    print(f'val[{a}]:{round(train_score, 3)}')
    
```

```python
# alpha = 0 gives the lowest RMSE
```
